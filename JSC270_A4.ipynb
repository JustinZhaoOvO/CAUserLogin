Cleanup Data

```python
# Define a function for data cleaning, further cleaning needs are also going to be added to this function
def clean_data(data):
  # remove observations with missing labels
  cleaned_data = data.dropna(subset=['Sentiment'])
  # remove observations with labels other than {0, 1, 2}
  cleaned_data = cleaned_data[cleaned_data['Sentiment'].isin(["0", "1", "2"])]

  return cleaned_data
```

```python
cleaned_twitter_train = clean_data(twitter_train)
cleaned_twitter_test = clean_data(twitter_test)
```

### A)

```python
# Calculate Proportion of observations in each class of sentiment labels
for sentiment in ["0", "1", "2"]:
  count = (cleaned_twitter_train['Sentiment'] == sentiment).sum()
  if count > 0:
    print(f'The proportion of sentiment {sentiment} is {count / len(cleaned_twitter_train) * 100}%')
  else:
    print(f"Sentiment {sentiment} has no observations")
```
